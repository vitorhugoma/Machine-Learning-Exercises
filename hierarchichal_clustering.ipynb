This code presents an hierarchichal clustering of a data set.

```python
#Import libraries

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
#Looking the data set
base = np.array([[2,10],[2,5],[8,4],[5,8],[7,5],
                 [6,4],[1,2],[4,9]])

for i in range(8):
  plt.scatter(base[i,0], base[i,1], label = 'A'+ str(i))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data set')
plt.legend() 
plt.show()
```

```python
#Dendogram plot and clustering operation

dendrograma = dendrogram(linkage(base, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Points')
plt.ylabel('Euclidean Distance')
plt.show()

hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
previsoes = hc.fit_predict(base)

plt.scatter(base[previsoes == 0, 0], base[previsoes == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(base[previsoes == 1, 0], base[previsoes == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(base[previsoes == 2, 0], base[previsoes == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Nº clusters: 3')
plt.legend()   
```
